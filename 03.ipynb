```python
import tensorflow as tf
import keras
import numpy as np
import matplotlib.pyplot as plt
```

```python
from keras.src import datasets, layers, models
```

```python
(train_images, train_labels), (test_images, test_labels) = tf.keras.datasets.fashion_mnist.load_data()
```

```python
print(train_images.shape)
```

```python
print(train_labels.shape)
```

```python
print(train_labels[0])
```

```python
train_images = train_images.reshape((60000, 28, 28)).astype('float32') / 255.0
test_images = test_images.reshape((10000, 28, 28)).astype('float32') / 255.0
```

```python
train_labels = tf.keras.utils.to_categorical(train_labels)
test_labels = tf.keras.utils.to_categorical(test_labels)
```

```python
model = keras.Sequential()
model.add(keras.layers.Input(shape=(28,28,)))
model.add(keras.layers.Flatten())
model.add(keras.layers.Dense(128, activation='relu'))
model.add(keras.layers.Dense(10, activation='softmax'))
model.summary()
```

```python
model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',  # 'sparse_'를 제거해야 합니다!
    metrics=['categorical_accuracy']
)
```

```python
model.fit(train_images, train_labels, epochs=5, batch_size=128)
```
